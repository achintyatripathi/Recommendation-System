```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
%matplotlib inline
```

```python
#movie.csv has the list of movies and there informations
movie_df = pd.read_csv('movies.csv')
# rating.csv has the ratings of all the movies 
rating_df = pd.read_csv('ratings.csv')

movie_df.head()
```

```python
# we need seperate the year from title .str.extract('(\d\d..\)')
```

```python
movie_df['year'] = movie_df.title.str.extract('(\(\d\d\d\d\))' ,expand=False)
movie_df.head()
```

```python
# now remove paranthesis from year
```

```python
movie_df['year'] = movie_df['year'].str.extract('(\d\d\d\d)' , expand=False)
movie_df.head()
```

```python
#remove year from title 
movie_df['title'] = movie_df.title.str.replace('(\(\d\d\d\d\))','')
#Applying the strip function to get rid of any ending whitespace characters that may have appeared
movie_df['title'] = movie_df['title'].apply(lambda x: x.strip())
movie_df.head()
```

```python
movie_df['genres'] = movie_df.genres.str.split('|')
movie_df.head()
```

```python
movieswithGenres_df = movie_df.copy()
# we use iterrows to go throw all the elements of the file. 
# then we use .at[index,gener] to incr. the movie if found in the gener part
for index,row in movie_df.iterrows():
    for gener in row['genres']:
        movieswithGenres_df.at[index,gener] = 1
# fillna all the generic movies with 0 where values are NaN
movieswithGenres_df = movieswithGenres_df.fillna(0)
movieswithGenres_df.head()

```

```python
rating_df.head()
```

```python
rating_df = rating_df.drop('timestamp',1)
rating_df.head()
```

# Content Based Recomendation

```python
# first we create a user-list  
```

```python
userInput = [
            {'title':'Breakfast Club, The', 'rating':5},
            {'title':'Toy Story', 'rating':3.5},
            {'title':'Jumanji', 'rating':2},
            {'title':"Pulp Fiction", 'rating':5},
            {'title':'Akira', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies['title']
```

```python
inputId = movie_df[movie_df['title'].isin(inputMovies['title'].tolist())]
#Then merging it so we can get the movieId. It's implicitly merging it by title.
inputMovies = pd.merge(inputId, inputMovies)
#Dropping information we won't use from the input dataframe
inputMovies = inputMovies.drop('genres', 1).drop('year', 1)
#Final input dataframe
#If a movie you added in above isn't here, then it might not be in the original 
#dataframe or it might spelled differently, please check capitalisation.
inputMovies
```

```python
#Filtering out the movies from the input
userMovies = movieswithGenres_df[movieswithGenres_df['movieId'].isin(inputMovies['movieId'].tolist())]
userMovies
```

```python
userMovies = userMovies.reset_index(drop =True)
userMovies = userMovies.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
userMovies
```

```python
inputMovies['rating']
```

```python
userprofile = userMovies.transpose().dot(inputMovies['rating'])
userprofile
```

 [[adventure],
    [Animation],
    [.....]]  *  [ratings]
    
    it give what type of genric movies does the user like more. 
    

```python
generTable =  movieswithGenres_df.set_index(movieswithGenres_df['movieId'])
generTable
```

```python
# now remove the unwanted material like movieid title gener year 

generTable = generTable.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
generTable.head()
```

```python
generTable.shape
```

generTable [N][M] * userprofile [M][1]
generTable contains information about geners the movie has
and 
userprofile contains a column with the % of the type of movies he like.

```python
recomendationTable_df = ((generTable*userprofile).sum(axis=1))/(userprofile.sum())
```

```python
recomendationTable_df.head()
```

```python
recomendationTable_df = recomendationTable_df.sort_values(ascending = False)
recomendationTable_df.head()
```

```python
# these can the gener movies the user my want to see.
```

```python
final_recomendation = movie_df.loc[movie_df['movieId'].isin(recomendationTable_df.head(20).keys())]

final_recomendation.reset_index(drop=True)
```

### Advantages and Disadvantages of Content-Based Filtering
#### Advantages
Learns user's preferences<br>
Highly personalized for the user
#### Disadvantages
Doesn't take into account what others think of the item, so low quality item recommendations might happen<br>
Extracting data is not always intuitive<br>
Determining what characteristics of the item the user dislikes or likes is not always obvious
